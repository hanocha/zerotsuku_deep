# 活性化関数

## ステップ関数の実装

```python
def step_function(x):
  return 1 if x > 0 else 0

print(step_function(-0.1))
```

numpy.array を受け取るようにする

```python
import numpy

def step_function_for_array(x):
  y = x > 0
  return y.astype(numpy.int)

test_array = numpy.array([0, 0.1, -0.1])
print(step_function_for_array(test_array))
```

## ステップ関数のグラフ

```python
import numpy as np
import matplotlib.pylab as plt

x = np.arange(-5.0, 5.0, 0.1)
y = step_function_for_array(x)

plt.plot(x, y)
plt.show()
```

## シグモイド関数

sigmoid(x) に numpy.array を渡してもそのままちゃんと計算が成立する :thinking:

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

sig = sigmoid(x)
plt.plot(x, sig)
plt.show()
```

## ReLU  関数

Rectified Linear Unit

```python
def relu(x):
  return np.maximum(0, x)

relu_y = relu(x)
plt.plot(x, relu_y)
plt.show()
```
